```python
#Question 1

l=list(range(1,21))
print("List:", l)

s=0
p=1
for i in l:
    if i%2==0:
        s+=i
    else:
        p*=i
print("Sum of the even numbers:",s)
print("Product of odd numbers:",p)

def mean(x):
    return sum(x)/len(x)

def median(x):
    n=len(x)
    new=sorted(x)
    if n%2==0:
        return (new[n//2-1]+new[(n//2)])/2
    else:
        return new[(n-1)//2]

def sd(x):
    return (sum([(i-m)**2 for i in l])/len(x))**0.5
    
m=mean(l)
print("Mean:",m)
print("Median:",median(l))
print("Standard Deviation:",round(sd(l),6))
```

```python
#Question 15
word = "listen"
words = ["enlist", "google", "inlets", "banana"]
print("Given words",words)

def isanagram(check,ref):
    for i in check:
        if check.count(i)!=ref.count(i):
            return 0
    return 1

new=[]
for curr in words:
    if isanagram(curr,word):
        new.append(curr) 
print(f"Anagrams of {word}:",new)
```

```python
#Question 16
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

a=["river",
    "galaxy",
    "melody",
    "whisper",
    "thunder",
    "meadow",
    "horizon",
    "canyon",
    "echo",
    "shadow",]
b=["breeze",
    "sunrise",
    "canyon",
    "thunder",
    "crystal",
    "lantern",
    "compass",
    "shadow",
    "melody"]
aset=set(a)
bset=set(b)
print("Union:",aset|bset)
print("Intersection:",aset&bset)
print("Symmetric Difference:",aset.symmetric_difference(bset))

venn=venn2([aset,bset], set_labels = ('Group A', 'Group B')) 
venn.get_label_by_id('11').set_text("\n".join(aset&bset))
venn.get_label_by_id('10').set_text("\n".join(aset-bset))
venn.get_label_by_id('01').set_text("\n".join(bset-aset))
plt.show()
```

```python
#Question 18
from matplotlib import pyplot as plt
f=open("text.txt",'r')

words=[]
wdict=dict()
for line in f.readlines():
    words.extend(list(map(str,line.lower().split())))

for i in range(len(words)):
    for character in words[i]:
        if 96<ord(character)<123:
            continue
        else:
            words[i]=words[i].replace(character,'')

for curr in words:
    if curr in wdict.keys():
        wdict[curr]+=1
    else:
        wdict[curr]=1

print("The most frequent word:",max(wdict.keys(),key=wdict.get))
mostfreq=sorted(wdict.keys(),key=wdict.get,reverse=True)[:10]
theircounts=sorted(wdict.values(),reverse=True)[:10]
plt.title("The Top 10 frequent words")
plt.bar(mostfreq,theircounts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```
